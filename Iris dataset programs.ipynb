```python
#Calculate  Petal width mean ,median ,SD,Variance  for the species setosa
t=df["Species"]=="Iris-setosa"
k=df[t]["PetalWidthCm"]
print("mean : ",k.mean())
print("median : ",k.median())
print("Standard Deviation : ",k.std())
print("variance : ",k.var())
```
