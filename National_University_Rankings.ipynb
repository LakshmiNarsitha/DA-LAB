```python
import pandas as pd
df=pd.read_csv("National_Universities_Rankings.csv")
df
```

```python
#1.	Find the universities in which undergraduate students were admitted.
t=df['Undergrad Enrollment']!=0
print("Universities with undergraduate enrollment :\n",df['Name'][t])
```

```python
#2.	List the states along with the cities in which universities located.
df[['Location','Name']]
```

```python
#3.	List the cities &universities under each state.
city=[]
state=[]
p=list(df['Location'])
for i in p:
  k=i.split(",")[1]
  m=i.split(",")[0]
  city.append(m)
  state.append(k)
df["city"]=city
df["state"]=state
df[['city','state','Name']]
```

```python
#4.	How many universities in each state have ranking<50? Print them along with ranking.
k=df[df['Rank']<50]
print("no of univerties have rank below 50 are : ", len(k))
k[['Name','Rank']]
```

```python
a=df[df["Rank"]<50].groupby(["state","Name"])["Rank"].apply(list)
p=a.reset_index(name="Rank")
p
```

```python
#5.	How many universities have both out-of state and in-state students?
k=df[df['In-state'].notnull()==True]
print("no of universities with both in and out state students are : ",len(k))
p=df[df['In-state'].notnull()==False]
print("no of universities with only out state students : ",len(p))
```

```python
#6.	How many universities have marginal difference <=$5000 in in-state &out-of state tuition fees
k=df['In-state'].str.replace(",","").str.split("$",expand=True).fillna(0)
t=k[1].astype('int64')
m=df['Tuition and fees'].str.replace(",","").str.split("$",expand=True).fillna(0)
p=m[1].astype('int64')
n=df[abs(t-p)<=5000]
print("no of universities are : ",len(n))
n
```

```python
#7.	List the universities having tuition fee>15000$ and rank between 120 to 170
df['tfee']=p
df[((df['tfee']>1500) & (df['Rank']>120) & (df['Rank']<170))]
```

```python
#8.	Find the campuses of universities located in different cities (multiple cities)
df[['name', 'unicity']] = df['Name'].str.split('--', 1, expand=True)
print(df[['name', 'unicity']])
df[df["name"].duplicated()==True]["name"]
```

```python
#9.	Mention the states where out-of state fee is more than in-state students. Print Minimum and Maximum fees.
df['ifee']=t
k=df['tfee']>df['ifee']
p=df[k]['tfee'].min()
q=df[k]['tfee'].max()
print("minimum fees : ",p)
print("maximum fees : ",q)
```

```python
#10.Find the cities locating top 100 universities.
k=((df['Rank']>1) & (df['Rank']<100))
df[k]['city']
```

```python
#11.Find universities with least no of undergraduate students &highest no of graduate students.
p=df['Undergrad Enrollment'].str.replace(",","").astype('int64').min()
k=df['Undergrad Enrollment'].str.replace(",","").astype('int64').max()
print(p)
print(k)
```

```python
#12.Identifying correlations between enrollment numbers and university rank
p=df['Undergrad Enrollment'].str.replace(",","").str.split("$",expand=True).fillna(0)
df['Ug enroll']=p.astype('int64')
k=df['Rank'].corr(df['Ug enroll'])
k
```
