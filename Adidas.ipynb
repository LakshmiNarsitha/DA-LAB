```python
import pandas as pd
df=pd.read_excel("Adidas_Dataset.xlsx")
df
```

```python
#1.	List all the retailers with retailer id.
df[["Retailer","Retailer ID"]].drop_duplicates()
```

```python
#2.	List all the retailers in every region.
p=df.groupby('Region')["Retailer"].unique()
p
```

```python
#3.	List the retailers in every city of a state.
p=df.groupby(["State","City"])["Retailer"].unique()
p
```

```python
#4.	List the products sold by the retailer in every city.
p=df.groupby(["Retailer","City"])["Product"].unique()
p
```

```python
#5.	Find the total sales of every retailer.
p=df.groupby(["Retailer"])["Total Sales"].sum()
p
```

```python
#6.	Find the total sales of the retailers in every city along with profit.
p=df.groupby(["City","Retailer"])[["Total Sales","Operating Profit"]].sum()
p
```

```python
#7.	Find the total sales &profit of each product sold by the retailer.
p=df.groupby(["Retailer","Product"])[["Total Sales","Operating Profit"]].sum()
p
```

```python
#8.	Find the units sold, total sales &profit of the products sold between the dates 
#1/1/2020 and 4/15/2020.
p=df["Invoice Date"].between("1/1/2020","4/15/2020")
k=df[p].groupby(["Product"])[["Units Sold","Total Sales","Operating Profit"]].sum()
k
```

```python
#9.	Find the no of units sold of each product by each retailer in every city.
p=df.groupby(["Retailer","Product","City"])["Units Sold"].sum()
p
```
