```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("only_road_accidents_data3.csv")
df
```

```python
#1.Find no. of accidents happened during day time & night time in Andhra Pradesh and show them using plot.
k=df[df["STATE/UT"]=="Andhra Pradesh"]
p=pd.DataFrame()
p["day"]=k["6-9 hrs (Day)"]+k["9-12 hrs (Day)"]+k["9-12 hrs (Day)"]+k["15-18 hrs (Day)"]
p["night"]=k["0-3 hrs. (Night)"]+k["3-6 hrs. (Night)"]+k["18-21 hrs (Night)"]+k["21-24 hrs (Night)"]
plt.plot(k["YEAR"],p["day"],"ro-",label="day")
plt.plot(k["YEAR"],p["night"],"bo-",label="night")
plt.legend()
plt.show()
```

```python
#2.Find total no. of accidents happened in all states and draw the plot to show the results.
k=df[["Total","STATE/UT"]]
p=k.groupby("STATE/UT")["Total"].sum()
p.plot.bar()
plt.show()
```

```python
#3.Find the accidents happened during early hours in south and display them using subplots.
south=['A & N Islands','Andhra Pradesh','Chhattisgarh','D & N Haveli','Goa',
   'Karnataka','Kerala','Lakshadweep','Maharashtra','Odisha','Tamil Nadu','Puducherry']
p=df.groupby('STATE/UT')[['0-3 hrs. (Night)','3-6 hrs. (Night)']].sum()
p.loc[south].plot.bar()
plt.show()

plt.subplot(1,2,1)
plt.title('0-3 hrs. (Night)')
plt.plot(p.loc[south]['0-3 hrs. (Night)'],"ro-")
plt.xticks(rotation=90)

plt.subplot(1,2,2)
plt.title('3-6 hrs. (Night)')
plt.plot(p.loc[south]['3-6 hrs. (Night)'],"go-")
plt.xticks(rotation=90)
plt.show()
```

```python
#5.Find the states with accidents more than average no. of accidents during day time & night time and show them using plots.
df['day1']=df[['6-9 hrs (Day)','9-12 hrs (Day)','12-15 hrs (Day)','15-18 hrs (Day)']].sum(axis=1)
df['night1']=df[['0-3 hrs. (Night)','3-6 hrs. (Night)','18-21 hrs (Night)','21-24 hrs (Night)']].sum(axis=1)

avg1=df['day1'].mean()
p=df.groupby('STATE/UT')['day1'].sum()
p=p.reset_index('STATE/UT')
d=p[p['day1']>avg1][['STATE/UT','day1']]

avg2=df['night1'].mean()
k=df.groupby('STATE/UT')['night1'].sum()
k=k.reset_index('STATE/UT')
n=k[k['night1']>avg2][['STATE/UT','night1']]

plt.plot(d['STATE/UT'],d['day1'],'o-'),plt.title("Day Time")
plt.xticks(rotation=90)
plt.show()

plt.plot(n['STATE/UT'],n['night1'],'o-'),plt.title("Night Time")
plt.xticks(rotation=90)
plt.show()
```

```python
#6.Find no. of accidents happened in each year and show them using plots.
k=df[["Total","YEAR"]]
p=k.groupby("YEAR")["Total"].sum()
plt.plot(p,"go-")
plt.show()
```

```python
#8.Find the maximum, minimum & average no. of accidents happened in each year and draw plot to show the results.
k=df[["Total","YEAR"]]
m1=k.groupby("YEAR")["Total"].min()
m2=k.groupby("YEAR")["Total"].max()
m3=k.groupby("YEAR")["Total"].mean()
df1=pd.DataFrame({"Min":m1,"Max":m2,"Avg":m3})
df1.plot(marker="o")
plt.show()
```

```python
#9.Find accidents happened in all states during 3-6 hours from 2001 to 2014 and draw plot to show the results.
k=df[["3-6 hrs. (Night)","YEAR"]]
p=k.groupby("YEAR").sum()
plt.plot(p,"ro-")
plt.show()
```
