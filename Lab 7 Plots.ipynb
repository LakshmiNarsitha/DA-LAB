```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("csd.csv")
df
```

```python
#1.	Read Total profit of all months and show it using a line plot
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.plot(df["month_number"],df["total_profit"],"r-")
plt.show()
```

```python
#2.	Get total profit of all months and show line plot with the following Style properties
plt.plot(df['month_number'],df['total_profit'],color='b',linestyle='dotted',
         marker='o',markerfacecolor='g',markersize=10)
plt.show()
```

```python
#3.	Read all product sales data and show it using a multiline plot
k=df["month_number"]
plt.plot(k,df["facecream"])
plt.plot(k,df["facewash"])
plt.plot(k,df["toothpaste"])
plt.plot(k,df["bathingsoap"])
plt.plot(k,df["shampoo"])
plt.plot(k,df["moisturizer"])
plt.show()
```

```python
#4.	Read toothpaste sales data of each month and show it using a scatter plot
plt.scatter(df["month_number"],df["toothpaste"])
plt.show()
```

```python
#5.	Read face cream and facewash product sales data and show it using the bar chart
index=df['month_number']
p=pd.DataFrame({'Face Cream':df['facecream'],'Face Wash':df['facewash']})
p.plot.bar()
plt.show()
```

```python
#6.	Read sales data of bathing soap of all months and show it using a bar chart.
df['bathingsoap'].plot.bar()
plt.show()
```

```python
#7.	Read the total profit of each month and show it using the histogram to see the most common profit ranges
df['total_profit'].hist(bins=15)
plt.show()
```

```python
#8.	Calculate total sale data for last year for each product and show it using a Pie chart
p=df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']]
sum1=p.sum()
plt.pie(sum1,labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.legend(loc="lower right")
plt.show()
```

```python
#9.	Read Bathing soap facewash of all months and display it using the Subplot.
plt.subplot(1,2,1)
plt.plot(df['month_number'],df['bathingsoap'])
plt.subplot(1,2,2)
plt.plot(df['month_number'],df['facewash'])
plt.show()
```

```python
#10	Read all product sales data and show it using the stack plot
p=df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']]
p.plot(kind='bar',stacked=True)
plt.show()
```
