```python
#1.	Combine the CSE & IT data and display the data
import pandas as pd
df1=pd.read_csv("CSE.csv")
df2=pd.read_csv("IT.csv")
df2.rename(columns={"PE":"Professional Elective"},inplace=True)
df=df1.append(df2,ignore_index=True)
df11=df.copy()
df
```

```python
df3=pd.read_csv("STUDENT.csv")
dfp=df.merge(df3)
dfp
```

```python
#2.	Display all CSE students’ marks along with personal information.
dfc=df1.merge(df3)
dfc
```

```python
#3.	Print all students Regd.No, Name and professional elective.
dfp[["Regd.No","Name","Professional Elective"]]
```

```python
#4.	Identify the students whose DAA marks are >18.
dfp[dfp["DAA"]>18][["Name","Regd.No"]]
```

```python
#5.	Display the names and EAMCET ranks of the students who got minimum 12 marks in all courses.
k=(dfp["AFL"]>=12) & (dfp["CN"]>12) & (dfp["DAA"]>12) & (dfp["OE"]>12) & (dfp["Professional Elective"]>12)
dfp[k][["Name","EAMCET RANK"]]
```

```python
#6.	Calculate mean value of all the subject’s marks.
k=dfp[["AFL","CN","DAA","OE","Professional Elective"]].mean()
k
```

```python
#7.	Display the names common in both CSE & IT along with Regd.No.
dfp[dfp["Name"].duplicated()][["Name","Regd.No"]]
```

```python
#8.	Fill the missing values of the data with  average marks of the subject of specific group.
dfp["AFL"]=dfp["AFL"].fillna(df["AFL"].mean())
dfp["CN"]=dfp["CN"].fillna(df["CN"].mean())
dfp["DAA"]=dfp["DAA"].fillna(df["DAA"].mean())
dfp["OE"]=dfp["OE"].fillna(df["OE"].mean())
dfp["Professional Elective"]=dfp["Professional Elective"].fillna(df["Professional Elective"].mean())
dfp
```

```python
#9.	Divide the students into 5 groups based on average marks.
dfp["avg"]=(dfp["CN"]+dfp["AFL"]+dfp["DAA"]+dfp["OE"]+dfp["Professional Elective"])/5
dfp["group"]=pd.qcut(dfp["avg"],5,labels=False)
dfp[["Regd.No","group"]]
```

```python
#10.Create  equal sized groups of students based on EAMCET Rank.
dfp["group1"]=pd.qcut(dfp["EAMCET RANK"],6,labels=False)
dfp[["Regd.No","EAMCET RANK","group1"]]
```
